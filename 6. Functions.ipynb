# **3. args and kwargs**
`*args and **kwargs` are special Python keywords that are used to pass the variable length of argument of a function

***args**

allows us to pass a variable number of **non-keyword arguments** to a function

```python
def multiply(a,b):
  return a*b
```

```python
multiply(2,3)
```

```python
# ekhon jodi amake amr manager eshe bole je jekono shonkhong input nao
# koto input idk but ja ee hbe multiply kore diye dibe
# eishob cases e args ashe where you don't know how much input is there
```

```python
def multiply(*args):
  product = 1

  for i in args:
    product = product*i

  return product
```

```python
multiply(1,2,3,4,5,6)
```

```python
#we can use any name instead of args
def multi(*mahin):
  prod = 1

  for i in mahin:
    prod = prod*i

  return prod
```

```python
multi(1,2,3,4,5,6)
```

```python
#args stores the input inside a tuple
def multiply(*args):
  product = 1

  for i in args:
    product = product*i

  print(args)
  return product
```

```python
multiply(1,2,3)
```

***kwargs**

allows us to pass a variable number of **keyword arguments** to a function


*   **keyword arguments** mean that they contain a key-value pair,like a Python dictionary



```python
def display(**kwargs):
  for(key,value) in kwargs.items():
    print(key, '->' ,value)
```

```python
display(india='delhi',srilanka='colombo')
```

**Points to be remembered while using `args and kwargs`**
*   Order of the arguments matter ( normal -> `*args` -> `**kwargs` )
*   The words args and kwargs are only a convention, you can use any name of you choice

### Lambda Function

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

image.png

```python
# x -> x^2
a = lambda x:x**2
a(2)
```

```python
# x,y -> x+y
b = lambda x,y: x+y
b(5,1)
```

#### Diff between lambda vs Normal Function

- No name
- lambda has no return value(infact,returns a function)
- lambda is written in 1 line
- not reusable

Then why use lambda functions?<br>
**They are used with HOF**


```python
# check if a string has 'a'
a = lambda s : 'a' in s
a('hello')
```

```python
# odd or even
b = lambda x :'even' if x % 2 == 0 else 'odd'
b(5)
```

### Higher Order Functions
Is that function in which returns another function

```python
def transform(f,L):
  output = []
  for i in L:
    output.append(f(i))

  print(output)
L = [1,2,3,4,5]
transform(lambda x:x**3,L)
```

### Map
Map 2 ta jinish expect kore
1. Lambda Function
2. List

map diye kono list er upor logic apply kora jay

```python
# square the items of a list
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# odd/even labelling of list items
L = [1,2,3,4,5]
list(map(lambda x: 'even' if x % 2 ==0 else 'odd',L))
```

```python
# fetch names from a list of dict
users = [
    {
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
        'name':'Nitish',
        'age':33,
        'gender':'male'
    },
    {
        'name':'Ankita',
        'age':50,
        'gender':'female'
    }
]
list(map(lambda users:users['gender'],users))
```

### Filter


```python
# numbers greater than 5
L = [3,4,5,6,7]
list(filter(lambda x: x>5,L))
```

```python
# fetch fruits starting with 'a'
fruits = ['apple','guava','cherry']

list(filter(lambda x:x.startswith('a'),fruits))
```
